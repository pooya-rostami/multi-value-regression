```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

## loading train data

```python
df_train = pd.read_csv('train_data.csv',index_col='Id')
df_train
```

## all the columns that is going to is in the output of this cell

```python
df_train.replace({'MSZoning': {'C (all)':'C'}},inplace=True)
for col in df_train.columns:
    if col not in ['MSSubClass', 'LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',
                   '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'GarageYrBlt', 'GarageArea', 'WoodDeckSF', 'EnclosedPorch', 'OpenPorchSF', 'LowQualFinSF',
                  '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'SalePrice', 'MoSold', 'YrSold']:
        tmp = df_train[col].tolist()
        tmp = set(tmp)
        obj1 = enumerate(tmp)
        tmp1 = list(obj1)
        tmp1[0]
        dic = {}
        for i in range(len(tmp1)):
            dic[tmp1[i][1]] = int(tmp1[i][0])
        print(f"this column is {col}")
        print(dic)
        print("-"*100)
```

## preprocess on train data

```python
for col in df_train.columns:
    if col not in ['MSSubClass', 'LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',
                   '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'GarageYrBlt', 'GarageArea', 'WoodDeckSF', 'EnclosedPorch', 'OpenPorchSF', 'LowQualFinSF',
                  '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'SalePrice', 'MoSold', 'YrSold']:
        tmp = df_train[col].tolist()
        tmp = set(tmp)
        obj1 = enumerate(tmp)
        tmp1 = list(obj1)
        tmp1[0]
        dic = {}
        for i in range(len(tmp1)):
            dic[tmp1[i][1]] = int(tmp1[i][0])
        df_train.replace({col: dic}, inplace=True)
df_train
```

## replace NaN with mean of column in train data

```python
null_columns=df_train.columns[df_train.isnull().any()]
print(null_columns)
for i in null_columns:
    print(df_train[df_train[i].isnull()][null_columns])
for i in null_columns:
    df_train[i].fillna((df_train[i].mean()), inplace=True)
null_columns=df_train.columns[df_train.isnull().any()]
print(null_columns)
```

## seperate sale price from train data

```python
y = df_train['SalePrice'].copy()
df_train = df_train.drop(columns=['SalePrice'])
df_train
```

## we can't normalize y because "TweedieRegressor(power=2, alpha=0.5, link='log')" only gets [0,+inf) 

```python
# range_y = y.max() - y.min()
# mean_y = y.mean()
# y = (y - mean_y) / range_y
# y.head()
```

## delete columns with more than 0.75 correlation to reduce number of unnecessary features

```python
corr_matrix = df_train.corr()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop_train = [column for column in upper.columns if any(upper[column] > 0.75)]
print(to_drop_train)
df_train = df_train.drop(df_train[to_drop_train], axis=1)
df_train
```

```python
x_train = (df_train - df_train.mean())/(df_train.max() - df_train.min())
x_train.head()
```

## loading test data

```python
df_test = pd.read_csv('test_data.csv', index_col='Id')
df_test
```

## preprocess on test data

```python
for col in df_test.columns:
    if col not in ['MSSubClass', 'LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',
                   '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'GarageYrBlt', 'GarageArea', 'WoodDeckSF', 'EnclosedPorch', 'OpenPorchSF', 'LowQualFinSF',
                  '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'SalePrice', 'MoSold', 'YrSold']:
        tmp = df_test[col].tolist()
        tmp = set(tmp)
        obj1 = enumerate(tmp)
        tmp1 = list(obj1)
        tmp1[0]
        dic = {}
        for i in range(len(tmp1)):
            dic[tmp1[i][1]] = int(tmp1[i][0])
        df_test.replace({col: dic}, inplace=True)
df_test
```

## replace NaN with mean of column in test data

```python
null_columns=df_test.columns[df_test.isnull().any()]
print(null_columns)
for i in null_columns:
    print(df_test[df_test[i].isnull()][null_columns])
for i in null_columns:
    df_test[i].fillna((df_test[i].mean()), inplace=True)
null_columns=df_test.columns[df_test.isnull().any()]
print(null_columns)
```

## delete columns with more than 0.75 correlation to reduce number of unnecessary features (same column as train data)

```python
df_test = df_test.drop(df_test[to_drop_train], axis=1)
df_test
```

```python
x_test = (df_test - df_test.mean())/(df_test.max() - df_test.min())
x_test.head()
```

```python
from sklearn.linear_model import TweedieRegressor
reg = TweedieRegressor(power=2, alpha=0.5, link='log')
reg.fit(x_train, y)
```
